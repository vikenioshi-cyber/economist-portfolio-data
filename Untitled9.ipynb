```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# --- 1. SIMULAÇÃO DE DADOS (Mindset de Auditoria PwC) ---
# Criando um cenário de 1.000 transações de cartões de crédito/maquininha
np.random.seed(42)
n_transactions = 1000

data = {
    'transaction_id': range(10001, 10001 + n_transactions),
    'customer_id': np.random.randint(1001, 1050, n_transactions),
    'amount': np.random.exponential(scale=2000, size=n_transactions), # Maioria valores baixos
    'transaction_date': [datetime(2026, 3, 1) + timedelta(minutes=i*15) for i in range(n_transactions)],
    'declared_income': np.random.choice([3000, 5000, 8000, 15000], n_transactions),
    'merchant_category': np.random.choice(['Retail', 'Services', 'Electronics', 'Gambling', 'Luxury'], n_transactions)
}

df = pd.DataFrame(data)

# Inserindo um comportamento suspeito manual (Outlier para o teste)
df.loc[0, 'amount'] = 45000  # Transação muito acima da renda
df.loc[1, 'merchant_category'] = 'Gambling'
df.loc[1, 'amount'] = 12000 # Valor alto em categoria de risco

# --- 2. LÓGICA DE DETECÇÃO DE RISCO (AML Rules) ---
def aml_risk_scoring(row):
    score = 0
    # Regra 1: Transação > 80% da renda declarada (Discrepância)
    if row['amount'] > (row['declared_income'] * 0.8):
        score += 50
    # Regra 2: Valor absoluto acima de R$ 10.000 (Gatilho COAF/BCB)
    if row['amount'] > 10000:
        score += 30
    # Regra 3: Categorias de alto risco (Ex: Gambling/Luxury)
    if row['merchant_category'] in ['Gambling', 'Luxury']:
        score += 20
    return score

df['risk_score'] = df.apply(aml_risk_scoring, axis=1)

# --- 3. RELATÓRIO DE MONITORAMENTO (SAR - Suspicious Activity Report) ---
high_risk_cases = df[df['risk_score'] >= 50].sort_values(by='risk_score', ascending=False)

print("--- CLOUDWALK COMPLIANCE MONITORING ---")
print(f"Total Transactions Scanned: {len(df)}")
print(f"Suspicious Alerts Identified: {len(high_risk_cases)}")
print("\nTop Cases for Manual Review:")
print(high_risk_cases[['customer_id', 'amount', 'declared_income', 'risk_score', 'merchant_category']].head())
```
